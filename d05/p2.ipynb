```python
import numpy as np
np.random.seed(42)

def r2(y_true, y_pred):
  # Вычисление общей суммы квадратов
  tss = np.sum((y_true - np.mean(y_true))**2)
  # Вычисление объясненной суммы квадратов
  ess = np.sum((y_true - y_pred)**2)
  # Вычисление коэффициента детерминации
  r2_score = 1 - ess / tss
  return r2_score
```
